```python
#  Enhanced Shopping Trends Visualizations

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Load dataset
df = pd.read_csv("shopping_trends (1).csv")
 
print(df.head()) 


print("\n Data Types:\n")
print(df.dtypes)

# Count duplicate rows
duplicate_count = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicate_count}")


# Show the count of null (missing) values in each column
null_counts = df.isnull().sum()

# Display only columns that have missing values
null_counts[null_counts > 0]
print(null_counts)

```

```python
# Set Seaborn style
sns.set(style="whitegrid")
plt.rcParams.update({'font.size': 12})


#  1. Bar Chart - Top Categories Purchased
plt.figure(figsize=(8,5))
sns.countplot(
    data=df,
    y='Category',
    order=df['Category'].value_counts().index,
    palette="viridis"
)
plt.title('Top Categories Purchased', fontsize=14)
plt.xlabel('Count')
plt.ylabel('Category')
plt.tight_layout()
plt.show()

#  2. Pie Chart - Gender Distribution
gender_counts = df['Gender'].value_counts()
colors = ['#00b4d8', '#ff6b6b']
plt.figure(figsize=(6,6))
plt.pie(
    gender_counts,
    labels=gender_counts.index,
    autopct='%1.1f%%',
    startangle=140,
    colors=colors,
    textprops={'fontsize': 12}
)
plt.title('Gender Distribution', fontsize=14)
plt.tight_layout()
plt.show()

#  3. Box Plot - Purchase Amount by Category
plt.figure(figsize=(10,6))
sns.boxplot(
    data=df,
    x='Category',
    y='Purchase Amount (USD)',
    palette="Set3"
)
plt.title('Purchase Amount by Category', fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

#  4. Histogram - Age Distribution
plt.figure(figsize=(8,5))
sns.histplot(
    df['Age'],
    bins=20,
    kde=True,
    color='#48cae4'
)
plt.title('Age Distribution of Customers', fontsize=14)
plt.xlabel('Age')
plt.ylabel('Count')
plt.tight_layout()
plt.show()

#  5. Scatter Plot - Review Rating vs Purchase Amount
plt.figure(figsize=(8,5))
sns.scatterplot(
    data=df,
    x='Review Rating',
    y='Purchase Amount (USD)',
    hue='Gender',
    palette={'Male': '#0077b6', 'Female': '#f77f00'},
    alpha=0.7,
    edgecolor='k'
)
plt.title('Review Rating vs Purchase Amount', fontsize=14)
plt.tight_layout()
plt.show()
```
